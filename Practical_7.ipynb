### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot `y1 = x^2` in the first subplot and `y2 = x^3` in the second.


```python
# Exercise 9
x = np.linspace(0, 10, 100)
y1 = x ** 2  
y2 = x ** 3  

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(5, 8))

# First subplot: y = x
ax1.plot(x, y1, color='g', label='y = x')
ax1.set_title('Quadratic Function (y = x)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid(True)
ax1.legend()

# Second subplot: y = x
ax2.plot(x, y2, color='blue', label='y = x', linewidth=2)
ax2.set_title('Cubic Function (y = x)', fontsize=12)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.grid(True)
ax2.legend()
plt.tight_layout()
plt.show()
```

### Exercise 15:
Change the style of the plot to `ggplot` and replot the sine wave from Exercise 4.


```python
# Exercise 15
  
plt.style.use('ggplot')
x = np.linspace(0, 2 * np.pi, 100)  # 0 to 2 (full sine cycle)
y = np.sin(x)

plt.plot(x, y, 
         color='g',  
         linewidth=2, 
         label='y = sin(x)',
         markersize=4)

plt.title('Sine Wave (ggplot Style)')
plt.xlabel('x (radians)')
plt.ylabel('sin(x)')
plt.show()
```

### Exercise 17:
Create a plot with multiple subplots in a 2x2 grid. Plot `y1 = x`, `y2 = x^2`, `y3 = x^3`, and `y4 = x^4`.


```python
# Exercise 17
x = np.linspace(0, 5, 100)  # x from 0 to 5
y1 = x       # Linear
y2 = x**2    # Quadratic
y3 = x**3    # Cubic
y4 = x**4    # Quartic

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Polynomial Functions Subplots', fontsize=16)

# Plot 1
axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('y = x')
axs[0, 0].grid(True, linestyle='--', alpha=0.6)

# Plot 2
axs[0, 1].plot(x, y2, color='green')
axs[0, 1].set_title('y = x')
axs[0, 1].grid(True)

# Plot 3
axs[1, 0].plot(x, y3, color='red')
axs[1, 0].set_title('y = x')
axs[1, 0].grid(True)

# Plot 4
axs[1, 1].plot(x, y4, color='purple')
axs[1, 1].set_title('y = x')
axs[1, 1].grid(True)

plt.show()
```

### Exercise 18:
Create a heatmap using a 2D array of random values with shape (5, 5).


```python
# Exercise 18
data = np.random.rand(5, 5)
plt.figure(figsize=(6, 6))
heatmap = plt.imshow(data, cmap='viridis')

plt.colorbar(heatmap, label='Value')
plt.xticks(np.arange(5), labels=[f'X{i+1}' for i in range(5)])
plt.yticks(np.arange(5), labels=[f'Y{i+1}' for i in range(5)])

# Add value annotations
for i in range(5):
    for j in range(5):
        plt.text(j, i, f'{data[i, j]:.2f}', 
                 ha='center', va='center', 
                 color='white' if data[i, j] > 0.5 else 'black')

plt.title('55 Heatmap of Random Values', pad=20)

plt.tight_layout()
plt.show()
```

### Exercise 19:
Add error bars to a bar plot with `categories = ['A', 'B', 'C', 'D']` and `values = [3, 7, 5, 9]`, assuming the error is 1 for each value.

Plot a sine wave using a dashed line and a cosine wave using a solid line on the same plot.
